```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
import seaborn as sns
```

```python
categories = ['Food', 'Travel', 'Fashion', 'Fitness', 'Music', 'Culture', 'Family', 'Health']
```

```python
n=200
```

```python
date=pd.date_range('01-01-2024',periods=n)
```

```python
categories_random=[random.choice(categories)for _ in range(n)]
```

```python
likes_random = np.random.randint(0, 10000, size=n)
```

```python
data = {'date': date,
       'categories': categories_random,
       'likes':likes_random}
```

```python
df=pd.DataFrame(data)
```

```python
df.set_index('date', inplace= True)
```

```python
df['categories'].describe()
```

```python
# Print the count of each category
print(df['categories'].value_counts())
```

```python
df.info()
```

```python
df.duplicated()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

```python
sns.histplot(data=df, x='likes')

# Add titles and labels
plt.title('Distribution of Likes')
plt.xlabel('Number of Likes')
plt.ylabel('Frequency')

# Display the plot
plt.show()
```

```python
sns.boxplot(data=df, x='categories', y='likes')
plt.show()
```

```python
mean_likes = df['likes'].mean()
print(f'Mean of Likes: {mean_likes}')
```
